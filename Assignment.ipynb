```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('Salaries.csv')
df
```

## Basic Data Exploration

```python
#identifying the number of rows and columns in the dataframe 
print(f"Number of rows = {df.shape[0]}")
print(f"Number of columns = {df.shape[1]}")
```

```python
#number of the non null rows and the data type of each column
df.info()
numerical_columns=['TotalPay', 'BasePay', 'OvertimePay', 'OtherPay', 'Benefits']
```

```python
#number of null rows in the data
df.isnull().sum()
```

## Descriptive Statistics 

```python
#mean,median,standard deviation,min,max of each column 
df.describe()
```

```python
#mode of each column iloc[0] to get only the first mode
mode=df.mode().iloc[0]
mode
```

```python
#range=max-min
salary_range=df['TotalPay'].max()-df['TotalPay'].min()
print(f'the range of the totalpay is : {salary_range}')
```

## Data Cleaning

```python
#dropped status colum and notes colum since both are empty
data=df.drop(columns={'Status','Notes'})

#benefits colum and basepay colum mode is 0 so i fill the null with 0
data.loc[data['BasePay'].isnull(), 'BasePay'] = 0
data.loc[data['Benefits'].isnull(), 'Benefits'] = 0

#overtimepay and otherpay is has a symmetric distribution so i fill the null with the mean 
data['OvertimePay'].fillna(data['OvertimePay'].mean(), inplace=True)
data['OtherPay'].fillna(data['OtherPay'].mean(), inplace=True)
data.isnull().sum()
```

## Basic Data Visualization

```python
#histogram to show the majority of the employees total pay falls in the range of 0 to 10000
plt.figure(figsize=(10, 6))
sns.histplot(data['TotalPay'], bins=30, kde=True, color='navy')
plt.title('Distribution of Total Pay')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
#the pie chart shows the percentage of employees across different departments with the highest being the transit operator
plt.figure(figsize=(8, 8))
department_counts = data['JobTitle'].value_counts().head(10)
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

## Grouped Analysis

```python
#the total pay mean for each year
pay_peryear = data.groupby('Year')['TotalPay'].mean()
pay_peryear
```

```python
#the total pay mean according to the job title in each year 
jobpay_peryear = data.groupby(['Year', 'JobTitle'])['TotalPay'].mean()
jobpay_peryear
```

```python
#the highest paying job
highest_paying_job = jobpay_peryear.idxmax()
print(f"the highest paying job is {highest_paying_job[1]} in {highest_paying_job[0]} with a value of {jobpay_peryear.loc[highest_paying_job]}")
```

## Simple Correlation Analysis

```python
#correlation matrix to know the correlation between each numerical column
correlation_matrix = data[numerical_columns].corr()
correlation_matrix
```

```python
#heatmap to visualize the correlation between each numerical column
plt.figure(figsize=(10, 8))
sns.heatmap(data[numerical_columns].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#pair plot to identify clusters and potential relationships between different columns like the linear relationship between the total pay and base pay
sns.pairplot(data[numerical_columns])
plt.show()
```
